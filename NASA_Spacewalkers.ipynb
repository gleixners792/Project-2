```python
import pandas as pd
```

```python
csv_file = "spacewalkers.csv"

spacewalkers_df = pd.read_csv(csv_file, encoding="unicode_escape")
spacewalkers_df
```

```python
for col in spacewalkers_df.columns:
    print(col)
```

```python
spacewalkers_df.drop('Unnamed: 6', axis=1, inplace=True)
spacewalkers_df.drop('Unnamed: 7', axis=1, inplace=True)                
spacewalkers_df
```

```python
spacewalkers_df.set_index("Order")
spacewalkers_df
```

```python
spacewalkers_df.drop('Image', axis=1, inplace=True)
spacewalkers_df
```

```python
#(df.Name.apply(lambda x: pd.Series(str(x).split("_")))) 
spacewalkers_df[['Name', 'years']] = spacewalkers_df.Name.str.split('\r\n', expand=True)
spacewalkers_df
```

```python
spacewalkers_df.drop('years', axis=1, inplace=True)
spacewalkers_df
```

```python
spacewalkers_df['year'] = spacewalkers_df['Spacewalks'].str.extract('(\d+)').astype(float)
spacewalkers_df['Mission'] = spacewalkers_df['Spacewalks'].str.split('(', 1).str[0].str.strip()
spacewalkers_df

#df['weekNo'] = df['Financial Week Name'].str.extract('(\d+)').astype(int)
#df['titles1'] = df['movie_title'].str.split('(', 1).str[0].str.strip()
```

```python
spacewalkers_df[["year", "Mission"]] = spacewalkers_df['Spacewalks'].str.split(':', expand=True)
spacewalkers_df
```
